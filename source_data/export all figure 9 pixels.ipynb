```python
import numpy as np, pandas as pd
import PIL.Image
import os
np.set_printoptions(suppress=True)
```

```python
os.getcwd()
```

```python
y_max = 1400.0           # maximum y value (power)
y_start = 439.0 - 0.5    # pixel index for y=0.0
y_end = 19.0 - 0.5       # pixel index for y=y_max
x_start = 107.0 - 0.5
x_end = 1031.0 - 0.5
x_max = 167.0            # maximum x value (hour number)

plants = ["Firm RE", "AES", "Kalaeloa", "Kahe", "Waiau", "Cycling", "Peaking"]
powers = dict()
prev_total = None
for p in plants:
    f = os.path.join('Hawaii RPS Study - Figure 9', p + '.png')
    data = np.asarray(PIL.Image.open(f))
    rows, cols = data[:,:,3].nonzero()
    row_vals = np.empty(data.shape[:-1])
    row_vals[:] = np.nan
    row_vals[rows, cols] = rows
    #with np.errstate(all='ignore'):
    total_power = (np.nanmean(row_vals, axis=0) - y_start) * y_max / (y_end-y_start)
    if prev_total is None:
        prev_total = np.zeros(total_power.shape)
    power = np.nan_to_num(total_power - prev_total)
    prev_total += power
    powers[p] = power
powers
```

```python
data.shape
```

```python
df = pd.DataFrame(
    powers, 
    index=(np.arange(len(prev_total)) - x_start) * x_max / (x_end-x_start),
    columns=plants
)
df.iloc[500:600]
```

```python
df.loc[0:167].to_csv('figure 9 all pixels.csv')
```
